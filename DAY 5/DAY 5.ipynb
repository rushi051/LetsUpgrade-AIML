# QUE 3

```python
def isPalindrome(s): 
    return s == s[::-1] 
  
  

s = "malayalam"
ans = isPalindrome(s) 
  
if ans: 
    print("This string is a Palindrome") 
else: 
    print("This string is not a Palindrome") 

    
    
```
